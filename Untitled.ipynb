```python
plt.figure(figsize=(10,13))
df['Region or State'].head().value_counts().plot.pie(autopct="%1.1f%%")
# note use of pie diagram
plt.title('Percentage',fontsize=25)
# not labeling of diagram
plt.show()
```

```python
df.head(2)
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="Acre Value", data=df, order=df['Acre Value'].head(20).value_counts().index)
# note here we are using seaborn as sns which is an in-nuilt library in python that uses matplotlab underneath to plot graphs.
plt.xticks(rotation='vertical')
plt.title('Arce Value By Region/State', fontsize=25) # note labeling the df
plt.show()
```
